```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load the dataset (use read_excel for .xlsx files)
df = pd.read_excel('income.xlsx')

# Display the column names and the first few rows to inspect the data
print("Column names:", df.columns)
print("Data preview:")
print(df.head())

# Clean column names by stripping any leading/trailing spaces and standardizing the case
df.columns = df.columns.str.strip().str.lower()  # Remove spaces and convert to lowercase

# Now check the column names after cleaning
print("Cleaned column names:", df.columns)

# Ensure the columns 'age' and 'income' exist
if 'age' not in df.columns or 'income' not in df.columns:
    raise ValueError("The dataset does not contain 'age' and 'income' columns")

# Preprocessing: Scaling the data (since K-Means is sensitive to the scale of features)
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[['age', 'income']])

# Apply K-Means with 3 clusters
km = KMeans(n_clusters=3, random_state=42)
y_predicted = km.fit_predict(df_scaled)

# Add the cluster assignments to the dataframe
df['cluster'] = y_predicted

# Display the cluster centers (centroids)
print("Cluster Centers: \n", km.cluster_centers_)

# Plotting the clusters
plt.scatter(df['age'], df['income'], c=df['cluster'], cmap='viridis')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.title('K-Means Clustering (3 clusters)')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Dataset with the X, Y, and Z coordinates for the 8 objects
data = {
    'Objects': ['OB-1', 'OB-2', 'OB-3', 'OB-4', 'OB-5', 'OB-6', 'OB-7', 'OB-8'],
    'X': [1, 1, 1, 2, 1, 2, 1, 2],
    'Y': [4, 2, 4, 1, 1, 4, 1, 1],
    'Z': [1, 2, 2, 2, 1, 2, 2, 1]
}

# Creating the DataFrame
df = pd.DataFrame(data)

# Define the number of clusters (K=2)
kmeans = KMeans(n_clusters=2, random_state=42)

# Apply K-Means clustering on the 'X', 'Y', and 'Z' coordinates
df['Cluster'] = kmeans.fit_predict(df[['X', 'Y', 'Z']])

# Get the centroids of the clusters
centroids = kmeans.cluster_centers_

# Print the cluster centers (centroids)
print("Cluster Centers (X, Y, Z):")
print(centroids)

# Print the dataset with assigned cluster labels
print("\nDataset with cluster labels:")
print(df)

# Plotting the clusters (using only X and Y for 2D visualization)
plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='viridis')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], color='red', marker='X', s=200, label='Centroids')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-Means Clustering (K=2)')
plt.legend()
plt.show()
```
